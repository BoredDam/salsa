Giorgino sta giocando da solo al Gioco dell’oca. Ad ogni turno lancia un dado a sei facce e si muove di tante 
caselle quante indicate dal dado. Il gioco finisce quando Giorgino ritorna al via dopo aver attraversato tutto
il tabellone.

*a) Sia X la variabile aleatoria che desrive l’evento ”lancio del dado”. Che distribuzione segue X?*

*b) Assumendo che il dado sia equilibrato e che il tabellone abbia 100 caselle, simulare una partita al gioco dell’oca. Plottare i risultati ottenuti ad ogni lancio in funzione del numero del lancio. (hint: ricordate che il gioco si arresta quando si supera la linea del via!)*

*c) Qual è il risultato più frequente? Qual è il risultato meno frequente? Qual è la media? Commentare se i dati rispecchiano i risultati che ci aspettiamo dalla teoria.*

*d) Quanti lanci ha impiegato Giorgino per completare la prima partita? Ripetere la simulazione per 100 partite e calcolare il numero medio di lanci necessario per finire il gioco. (hint: ogni partita si concluderà in un numero diverso di mosse. Il numero di mosse è una v.a.: posso calcolarne la media!)*



```python
# Esercizio A

from scipy.stats import randint

# il dado non truccato segue una distribuzione uniforme discreta.
for i in range(0,8):
    p = randint.pmf(i, 1, 7)
    print(f"P(result={i}): {p}")
```

```python
# Esercizio B
import numpy as np
import matplotlib.pyplot as plt

sum = 0
rolls = []

while sum < 100:
    r = np.random.randint(0,7,1)[0]
    sum += r
    rolls.append(r)

plt.plot(range(len(rolls)), rolls)

plt.ylabel("dice roll")
plt.xlabel("turn")

plt.show()
```

```python
# Esercizio C

counts = [0]*6

for r in rolls:
    counts[r-1] += 1

max_c = counts.index(np.max(counts))+1
min_c = counts.index(np.min(counts))+1

print(f"Il risultato più frequente è stato {max_c}")
print(f"Il risultato meno frequente è stato {min_c}")

sum_rolls = np.sum(rolls)
mean = sum_rolls/len(rolls)
print(f"La media dei tiri è: {mean}")

plt.plot(range(len(rolls)), rolls)
plt.plot([0,len(rolls)], [mean, mean], '--', color='black', label='mean')
plt.legend()
plt.ylabel("dice roll")
plt.xlabel("turn")

plt.show()
```

```python
# Esercizio D

def game_sim(slots):
    i = 0
    sum = 0
    while sum < slots:
        r = np.random.randint(0,7,1)[0]
        sum += r
        i += 1
    return i

iterations = 100
rolls_amount = []
for i in range(iterations): 
    c = game_sim(100)
    rolls_amount.append(c)

mean = np.mean(rolls_amount)
print(f"Il numero medio di tiri richiesti per finire una partita è {mean}")
```
