```
a) Scrivere un codice che simuli N lanci di una moneta assumendo p = 0.5.
b) Plottare gli istogrammi per N = 100, 1000, 10000 e confrontarli con la distribuzione teorica attesa.
c) Ripetere l’esperimento nel caso di moneta truccata con p = 0 − 7.
d) Commentare i risultati ottenuti.
```

````python
import numpy as np
````

````python
# supponiamo testa = 0

shapes = [100,1000,10000]
exp = []
for s in shapes:
    coins = np.random.randint(0,2, [s])
    counts = np.count_nonzero(coins)/s
    exp.append(counts)
    print(counts)
````

````python
#!pip install matplotlib

import matplotlib.pyplot as plt

for c, e in enumerate(exp):
    plt.bar([0,1], [0.5,0.5], width=1)
    plt.bar([0,1], [e, 1-e])
    plt.title(shapes[c])
    plt.xticks([0,1])
    
    plt.show()
````

````python
# supponiamo testa = 0

shapes = [100,1000,10000]
exp = []
for s in shapes:
    coins = np.random.random([s]) > 0.7
    counts = np.count_nonzero(coins)/s
    exp.append(counts)
    print(counts)
````

````python
for c, e in enumerate(exp):
    plt.bar([0,1], [0.5,0.5], width=1)
    plt.bar([0,1], [e, 1-e])
    plt.title(shapes[c])
    plt.xticks([0,1])
    plt.show()
````

````python
for c, e in enumerate(exp):
    plt.bar([0,1], [0.3,0.7], width=1)
    plt.bar([0,1], [e, 1-e])
    plt.title(shapes[c])
    plt.xticks([0,1])
    plt.show()
````

In entrambi i casi il crescere del numero di monete su cui si effettua l'esperiumento, la distribuzione reale si avvicina a quella attesa.

```
a) Calcolare utilizzando il metodo hit or miss il seguente integrale definito:
b) Plottare il risultato dell’integrale in funzione del numero di punti generati.
c) Ripetere il conto nel caso in cui a = −1 e b = 1.
d) Quale converge pi`u rapidamente?
```

````python
def foo(x):
    return np.exp((-x**2)/2) * 1/np.sqrt(2 * np.pi)
````

````python
N = range(100, 10000, 200)
a = -3
b = 3
h = 1

plot = []

for n in N:
    res_x = np.random.uniform(a, b, n)
    res_y = np.random.uniform(0, 1, n)
    hit = 0

    for i in range(n):
        if res_y[i] <= foo(res_x[i]):
            hit += 1

    ratio = hit/n

    base = b*(n-1) - a
    area = base * h * ratio
    
    plot.append(area)

plt.plot(N,plot)
plt.show()
````
