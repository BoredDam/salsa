## Quesito 1


Uno studente ha notato che la distribuzione della scelta dei pesi della lat machine segue una distribuzione normale con media 45kg. Decide di indagare più a fondo facendo un’indagine statistica tra e persone fre- quentanti la sua palestra, la palestra “Muscled Mario”, e un’altra palestra, la “Weak Luigi”. Nella seguente tabella ci sono i risultati: nella prima riga c’è l’elenco dei chili sollevati alla lat, nelle altre due il numero di persone che ha dichiarato di sollevare quel peso nella palestra “Muscled Mario” e nella palestra “Weak Luigi”.


| Pesi Lat | 15 | 20 | 25 | 30 | 35 | 40 | 45 | 50 | 55 | 60 | 65 | 70 | 75 | 80 |
|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|
|*Muscled Mario*|0|1|2|20|34|50|64|59|40|18|11|1|0|0|
|*Weak Luigi*|1|6|19|23|52|75|82|93|63|44|21|10|8|3|

*a) Graficare in due istogrammi i dati raccolti dai due campioni.*

*b) Ci si aspetta che i dati provenienti dalla palestra “Muscled Mario” siano estratti da una popolazione con varianza σ 2 = 5, mentre quello estratto dalla palestra “Weak Luigi” con varianza σ 2 = 3. Trovare uno stimatore della media.*

*c) Calcolare l’intervallo di confidenza della media con α = 0.05.*

*d) Possiamo affermare con confidenza al 5% che la media dei pesi nella palestra “Muscled Mario” sia più bassa di quella dei pesi nella palestra “Weak Luigi”?*



```python
# Esercizio A

import numpy as np
import matplotlib.pyplot as plt

X = np.array(range(15, 81, 5))
M = np.array([0,1,2,20,34,50,64,59,40,18,11,1,0,0])
L = np.array([1,6,19,23,52,75,82,93,63,44,21,10,8,3])

plt.bar(X, M/np.sum(M), color='red', width=2)
plt.bar(X, L/np.sum(L), color='green', width=1.1)
plt.show()
```

```python
# Esercizio B

var_M = 5
var_L = 3

mean_M = np.sum(X*M)/np.sum(M)
mean_L = np.sum(X*L)/np.sum(L)

print(f"Media da Muscle Mario: {mean_M}kg\nMedia da Weak Luigi: {mean_L}kg")
```

```python
# Esercizio C
from scipy.stats import norm

alpha = 0.05
n_M = np.sum(M)
n_L = np.sum(L)

I_lower_M = mean_M - norm.ppf(1-alpha)*var_M/np.sqrt(n_M)
I_upper_M = mean_M + norm.ppf(1-alpha)*var_M/np.sqrt(n_M)
I_lower_L = mean_L - norm.ppf(1-alpha)*var_L/np.sqrt(n_L)
I_upper_L = mean_L + norm.ppf(1-alpha)*var_L/np.sqrt(n_L)

print(f"Intervalli di confidenza della media con alpha=0.05\n\tMuscle Mario: {float(I_lower_M), float(I_upper_M)}\n\tMedia da Weak Luigi: {float(I_lower_L), float(I_upper_L)}")
```
