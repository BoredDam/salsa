```python
import numpy as np
from scipy.stats import t

X = np.array([1.3, 2.7, 2.5, 4.1, 2.4, 2.8, 3.5, 1.2, 1.9, 2.6])
alpha = 0.01
n = len(X)
mean = np.mean(X)
var = np.sum((X - mean)**2)/(n-1)

t = t.ppf(q= 1-alpha/2, df=n-1)
I_0 = mean - np.sqrt(var/n)*t
I_1 = mean + np.sqrt(var/n)*t

I_0 = float(I_0)
I_1 = float(I_1)

print(f"numero di campioni: {n}")
print(f"media campionaria: {mean}")
print(f"varianza campionaria: {var}")
print(f"intervallo di confidenza: [{I_0,I_1}]")
```
